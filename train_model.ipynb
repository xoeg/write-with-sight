```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

Generate training sets and validation sets

```python
datadir = 'Data_set'
train_ds = tf.keras.utils.image_dataset_from_directory(
    datadir,
    image_size = (480, 640),
    validation_split = 0.2,
    color_mode = 'grayscale',
    subset = "training",
    seed = 123)

valid_ds = tf.keras.utils.image_dataset_from_directory(
    datadir,
    image_size = (480, 640),
    validation_split = 0.2,
    color_mode = 'grayscale',
    subset = "validation",
    seed = 123)
```

Check some of the images

```python
for image_batch, labels_batch in train_ds:
    Im = image_batch[1].numpy().astype("uint8")
    plt.imshow(Im)
    break
```

We need to crop the image so only the relevant features are processed by the CNN. In this case this corresponds to the eyesight. So we crop the image to ensure the eyes are on the image.

Define the class names

```python
class_names = train_ds.class_names
print(class_names)
```

Test some of the images and their labels:

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

Build the CNN, first layers are preprocessing of the image (cropping and rescaling). Next comes the CNN and finaly a set of dense layers. 

```python
def build_model():
    model = tf.keras.models.Sequential([
        # Preprocessing
        tf.keras.layers.Resizing(height = 160, width = 215, input_shape = (480,640,1)),
        tf.keras.layers.Rescaling(1./255),
        # CNN
        tf.keras.layers.Conv2D(16, (3,3), activation = 'relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(16, (3,3), activation = 'relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(16, (3,3), activation = 'relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(16, (3,3), activation = 'relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Flatten(),
        # Dense NN
        tf.keras.layers.Dense(64, activation = 'relu'),
        tf.keras.layers.Dense(64, activation = 'relu'),
        tf.keras.layers.Dense(20, activation = 'softmax')
        ])
    
    model.compile(loss = tf.losses.SparseCategoricalCrossentropy(),
                  optimizer = 'adam',
                  metrics = ['accuracy'])
    return model

model = build_model()
model.summary()
```

Train the model

```python
history = model.fit(train_ds, 
                    epochs = 100,
                    validation_data = valid_ds,
                    verbose = 1)
```

Check learning curves

```python
epochs = range(len(history.history['loss']))
f, fig = plt.subplots(1,2, figsize=(10,2))
fig[0].plot  ( epochs,     history.history[     'loss' ], label='Training')
fig[0].plot  ( epochs, history.history[     'val_loss' ], label='Validation')
plt.title ('Training and validation loss')
fig[0].legend()
plt.xlabel('epochs')

fig[1].plot  ( epochs,     history.history[     'accuracy' ], label='Training')
fig[1].plot  ( epochs, history.history[     'val_accuracy' ], label='Validation')
plt.legend()
plt.title ('Training and validation accuracy')
plt.xlabel('epochs')
```

Test some predictions

```python
for image_batch, labels_batch in valid_ds:
    x = image_batch[1]
    y = labels_batch[1]
    image = x 
    plt.imshow(image)
    print('Label: ', int(y), '=>',class_names[int(y)])
    pred = np.argmax(model.predict(np.expand_dims(x, axis=0)));
    print('Prediction: ',pred, '=>',class_names[int(pred)])
    break
```

Save model to use in the program

```python
model.save('Models/eye_tracker')
```
